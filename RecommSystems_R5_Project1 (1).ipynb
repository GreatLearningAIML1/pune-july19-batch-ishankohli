# Build Collaborative Filtering model and Finding RMSE

# 1. item-item collab filtering

```python
# Use user_based true/false to switch between user-based or item-based collaborative filtering
algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': False})
algo.fit(trainset)
```

```python
test_pred = algo.test(testset)
```

```python
test_pred
```

```python
# get RMSE
print("Item-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

# 2. User - User Collab Filtering

```python
algo1 = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo1.fit(trainset)
```

```python
test_pred1 = algo1.test(testset)
```

# Get top - K ( K = 5) recommendations. Since our goal is to recommend new products to each user based on his/her habits, we will recommend 5 new products.

```python
## uSing SVD
def get_top_n(predictions, n):
    # First map the predictions to each user.
    top_n = collections.defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
top_n = get_top_n(predictions, n=5)
```

```python
top_n
```

```python
# Print the recommended items for each user
for uid, user_ratings in top_n.items():
    print(uid, [iid for (iid, _) in user_ratings])
```

1) The RMSE is the square root of the variance of the residuals. It indicates the absolute fit of the model to the datahow close the observed data points are to the models predicted values. Whereas R-squared is a relative measure of fit, RMSE is an absolute measure of fit
2) Lower values of RMSE indicate better fit.
3) RMSE of ITEM-ITEM :- 1.1158 , USER-USER:- 1.0428 and SVD :- 1.0017
